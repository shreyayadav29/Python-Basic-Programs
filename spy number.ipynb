## spy number

### A Spy Number is a number in which the sum of its digits is equal to the product of its digits.


### Examples

1124

Sum = 1 + 1 + 2 + 4 = 8
Product = 1 × 1 × 2 × 4 = 8
✅ 1124 is a Spy Number

123

Sum = 1 + 2 + 3 = 6
Product = 1 × 2 × 3 = 6
✅ 123 is a Spy Number

145

Sum = 1 + 4 + 5 = 10
Product = 1 × 4 × 5 = 20
❌ 145 is not a Spy Number

```python
def sumDigits(n):
    sum=0
    while n!=0:
        sum=sum+(n%10)
        n=n//10
    return sum
def prodDigits(n):
    prod=1
    while n!=0:
        prod=prod*(n%10)
        n=n//10
    return prod
def spy(n):
    if sumDigits(n)==prodDigits(n):
        return True
    return False
n=int(input())
print(spy(n))
```
