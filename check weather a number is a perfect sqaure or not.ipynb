### Check Whether a Number is a Perfect Square or Not

A Perfect Square is a number that can be written as the square of an integer.

Examples:

4 = 2 × 2 ✅
9 = 3 × 3 ✅
16 = 4 × 4 ✅
20 ❌ (Not a perfect square)

```python
def perfectSquare(n):
    sq=n**0.5
    if sq==int(sq):
        return True
    return False
n=int(input())
print(perfectSquare(n))
```
