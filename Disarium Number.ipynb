## Amstrong Number

```python
def countDigits(n):
    count=0
    while n!=0:
        n=n//10
        count=count+1
    return count
def amstrong(n,c):
    copy_n=n
    sum=0
    while n!=0:
        sum=sum+(n%10)**c
        n=n//10
    if sum==copy_n:
        return True
    return False
n=int(input())
c=countDigits(n)
print(amstrong(n,c))
```
