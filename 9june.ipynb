## factorial of a number using iteration

```python
def factorial(n):
    fact=1
    while(n!=0):
        fact=fact*n
        n=n-1
    return fact
n=int(input())
print(factorial(n))
```

## factorial of a number using recursion

```python
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)
n=int(input())
print(factorial(n))
```

## Sum of Digits of a number

### eg. 532 , output= 10


```python
def sumDigits(n):
    sum=0
    while(n!=0):
        sum=sum+(n%10)
        n=n//10
    return sum
n=int(input())
print(sumDigits(n))
```

## Reversing a number

### eg. 6784 output= 4876

```python
def reverseNumber(n):
    sum=0
    while(n!=0):
        sum=(sum*10)+(n%10)
        n=n//10
    return sum
n=int(input())
print(reverseNumber(n))
```

## Palindrome Number

### eg. 121 , reverse=121

```python
def reverseNumber(n):
    sum=0
    while(n!=0):
        sum=(sum*10)+(n%10)
        n=n//10
    return sum
def palindrome(n):
    reverse=reverseNumber(n)
    if n==reverse:
        return True
    else:
        return False
n=int(input())
print(palindrome(n))
```
