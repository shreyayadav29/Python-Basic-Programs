## Prime Number

### A Prime Number is a positive integer that has exactly two factors: 1 and itself.

Examples
2 → Factors: 1, 2 ✅ Prime
3 → Factors: 1, 3 ✅ Prime
5 → Factors: 1, 5 ✅ Prime
7 → Factors: 1, 7 ✅ Prime
11 → Factors: 1, 11 ✅ Prime

Non-Prime (Composite) Example

12

Factors: 1, 2, 3, 4, 6, 12
More than two factors ❌ Not a Prime Number

```python
def prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=int(input())
print(prime(n))
```
