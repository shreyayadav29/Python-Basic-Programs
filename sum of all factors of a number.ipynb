## sum of all factors of a number

### The sum of all the factors of a number means adding every number that divides the given number exactly (with no remainder).

Example 1: Number = 12

Factors of 12 are:

1, 2, 3, 4, 6, 12

Sum of factors:

1+2+3+4+6+12=28

Output: 28

```python
def sumFactors(n):
    sum=0
    for i in range(1,n+1):
        if n%i==0:
            sum=sum+i
    return sum
n=int(input())
print(sumFactors(n))
```
