## sum of n natural numbers

### The sum of the first n natural numbers is the total of all numbers from 1 to n.


#### Example

If n = 5:

1+2+3+4+5=15

✅ Output: 15

```python
def sum_Natural(n):
    sum=0
    for i in range(1,n+1,1):
        sum=sum+i
    return sum
n=int(input())
print(sum_Natural(n))
```
