## Perfect Number

### A Perfect Number is a positive integer whose sum of all its proper factors (excluding the number itself) is equal to the number.
Example 1: Number = 6

Proper factors of 6 are:

1, 2, 3

Sum:

1+2+3=6

✅ 6 is a Perfect Number.

Example 2: Number = 28

Proper factors of 28 are:

1, 2, 4, 7, 14

Sum:

1+2+4+7+14=28

✅ 28 is a Perfect Number.

```python
def perfect(n):
    sum=0
    for i in range(1,n//2+1):
        if n%i==0:
            sum=sum+i
    if sum==n:
        return True
    return False
n=int(input())
print(perfect(n))
```
