## GCD/HCF of two numbers

### GCD / HCF of Two Numbers

The GCD (Greatest Common Divisor) or HCF (Highest Common Factor) is the largest positive number that divides both numbers exactly.

Example

Input: 24, 36

Factors of 24: 1, 2, 3, 4, 6, 8, 12, 24

Factors of 36: 1, 2, 3, 4, 6, 9, 12, 18, 36

Common factors: 1, 2, 3, 4, 6, 12

HCF = 12 ✅

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)
a=int(input())
b=int(input())
print(gcd(a,b))
```
