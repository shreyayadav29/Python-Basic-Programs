## strong number

### A Strong Number is a number whose sum of the factorials of its digits is equal to the number itself.
Examples

1!+4!+5!
=1+24+120
=145 âś… Strong Number

2!=2 âś… Strong Number

```python
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)
def strong(n):
    copy_n=n
    sum=0
    while n!=0:
        sum=sum+factorial(n%10)
        n=n//10
    if sum==copy_n:
        return True
    return False
n=int(input())
print(strong(n))
```
