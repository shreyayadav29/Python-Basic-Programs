## LCM of two numbers

```python
def lcm(a,b):
    maximum=max(a,b)
    i=maximum
    while(1):
        if maximum%a==0 and maximum%b==0:
            return maximum
        maximum=maximum+i
a=int(input())
b=int(input())
```
